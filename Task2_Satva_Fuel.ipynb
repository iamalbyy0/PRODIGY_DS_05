## Mean Fuel Consumption: 23.46 MT
## Standard Deviation: 0.54 MT
## Acceptable Range (±2σ): [22.37, 24.55] MT
## Satva Model Prediction: 24.3 MT

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Historical_fuel = [22.8, 23.1, 24.0, 23.5, 22.9, 23.7, 24.2]
Satva_Prediction = 24.3
```

```python
mean = np.mean(Historical_fuel)
std = np.std(Historical_fuel)
```

```python
print(mean)
print(std)
```

```python
print(f"Historical Fuel Consumption Data: {Historical_fuel}")
print(f"Satva Model Prediction: {Satva_Prediction} MT")
print(f"\n--- Statistical Analysis ---")
print(f"Mean: {mean:.4f} MT")
print(f"Standard Deviation: {std:.4f} MT")
```

```python
# Creating an Acceptable Range  
lower = mean - 2 * std
upper = mean +2 * std
```

```python
print(lower)
print(upper)
```

```python
pred = lower <= Satva_Prediction <= upper
pred
```

```python
if pred :
    print("Validation is acceptable")
else:
    print(" Validation is not in acceptable range")
```

```python
# Z-score
z_score = (pred - mean) / std

print("Mean:", round(mean,2))
print("Standard Deviation:", round(std,2))
print("Acceptable Range:", round(lower,2), "to", round(upper,2))
print("Z-score:", round(z_score,2))
```

```python
#  Historical vs Prediction 
plt.figure()
x = range(1, len(Historical_fuel)+1)
plt.scatter(x, Historical_fuel)
plt.axhline(mean)
plt.axhline(lower)
plt.axhline(upper)
plt.scatter(len(Historical_fuel)+1, pred, marker='*')
plt.title("Fuel Consumption Validation")
plt.xlabel("Voyage")
plt.ylabel("Fuel (MT)")
plt.show()
```
